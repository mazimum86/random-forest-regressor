## Random Forest

**Importing the Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Importing the dataset**

```python
data = pd.read_csv('position_salaries.csv')
data.head()
```

```python
X= data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
```

**Train the Random Forest Regression Model on the whole dataset**

```python
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor(n_estimators=10)
regressor_rf.fit(X,y)
```

**Predicting a new result**

```python
y_pred =regressor_rf.predict([[6.5]])
y_pred
```

**Visualising the Random Forest Regression Model results (for higher resolution and smoother curve)**

```python
X_ = np.linspace(X.min(),X.max(),10000).reshape(-1,1)
y_pred= regressor_rf.predict(X_)

plt.scatter(X,y,c='red')
plt.plot(X_,y_pred)
plt.title('Truth or Bluff (Random Forest)')
plt.xlabel('Experience')
plt.ylabel('Salary');
```
